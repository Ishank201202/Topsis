```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Result_102103035.csv")
```

```python

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
bar_positions = range(len(df['Model']))

ax.bar([pos - 1.5 * bar_width for pos in bar_positions], df['F1'], width=bar_width, label='F1-score')
ax.bar([pos - 0.5 * bar_width for pos in bar_positions], df['Acc'], width=bar_width, label='Accuracy')
ax.bar([pos + 0.5 * bar_width for pos in bar_positions], df['PPL'], width=bar_width, label='PPL')
ax.bar([pos + 1.5 * bar_width for pos in bar_positions], 100*df['TOPSIS Score'], width=bar_width, label='100x Topsis Score')

ax.set_xticks(bar_positions)
ax.set_xticklabels(df['Model'])
ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.legend()

plt.tight_layout()
plt.show()
```
